## IMPORTS

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

## Defining the expected function

```python
def func(x, m, c): return m*x + c
```

```python
xs = np.array([177.6, 155.4, 133.2, 111, 88.8, 66.6, 44.4, 22.2])[:3]
ys1 = np.array([235, 201, 166, 135, 132, 87, 58, 10.5])[:3]  # f = 500Hz 
ys2 = np.array([240, 203, 168, 138, 136, 86, 62, 19.7])[:3]  # f = 995Hz 
ys3 = np.array([246, 210, 171, 136, 133, 110, 63, 12])[:3]  # f = 1500Hz
ys4 = np.array([250, 214, 173, 136, 134, 110, 76, 14])[:3]  # f = 2000Hz
ys5 = np.array([255, 218, 172, 141, 137, 112, 88, 17])[:3]  # f = 2500Hz
ys6 = np.array([260, 223, 181, 146, 143, 108, 98, 26])[:3]  # f = 3000Hz
```

```python
# Fit the data
popt1, pcov1 = curve_fit(func, xs, ys1)  # f = 500Hz
popt2, pcov2 = curve_fit(func, xs, ys2)  # f = 995Hz
popt3, pcov3 = curve_fit(func, xs, ys3)  # f = 1500Hz
popt4, pcov4 = curve_fit(func, xs, ys4)  # f = 2000Hz
popt5, pcov5 = curve_fit(func, xs, ys5)  # f = 2500Hz
popt6, pcov6 = curve_fit(func, xs, ys6)  # f = 3000Hz
```

## Showing Results

```python
p = 5
text = "If the Equation is y = mi*x + ci\nFor the best fit curve:\n"
text += f"\tm1 = {round(popt1[0], p)} ± {round((pcov1[0][0])**0.5, p)},\t"
text += f"\tc1 = {round(popt1[1], p)} ± {round((pcov1[1][1])**0.5, p)};\t\n"
text += f"\tm2 = {round(popt2[0], p)} ± {round((pcov2[0][0])**0.5, p)},\t"
text += f"\tc2 = {round(popt2[1], p)} ± {round((pcov2[1][1])**0.5, p)};\t\n"
text += f"\tm3 = {round(popt3[0], p)} ± {round((pcov3[0][0])**0.5, p)},\t"
text += f"\tc3 = {round(popt3[1], p)} ± {round((pcov3[1][1])**0.5, p)};\t\n"
text += f"\tm4 = {round(popt4[0], p)} ± {round((pcov4[0][0])**0.5, p)},\t"
text += f"\tc4 = {round(popt4[1], p)} ± {round((pcov4[1][1])**0.5, p)};\t\n"
text += f"\tm5 = {round(popt5[0], p)} ± {round((pcov5[0][0])**0.5, p)},\t"
text += f"\tc5 = {round(popt5[1], p)} ± {round((pcov5[1][1])**0.5, p)};\t\n"
text += f"\tm6 = {round(popt6[0], p)} ± {round((pcov6[0][0])**0.5, p)},\t"
text += f"\tc6 = {round(popt6[1], p)} ± {round((pcov6[1][1])**0.5, p)}.\t"
print(text)
```

```python
factor = (max(xs) - min(xs))*0.05
xFit = np.arange(min(xs)-factor, max(xs)+factor, 0.01)

plt.figure(dpi=100)
plt.scatter(xs, ys1, c='r')
plt.plot(xFit, func(xFit, *popt1), "r", label="f = 500Hz ")
plt.scatter(xs, ys2, c='g')
plt.plot(xFit, func(xFit, *popt2), "g", label="f = 995Hz ")
plt.scatter(xs, ys3, c='b')
plt.plot(xFit, func(xFit, *popt3), "b", label="f = 1500Hz")
plt.scatter(xs, ys4, c='c')
plt.plot(xFit, func(xFit, *popt4), "c", label="f = 2000Hz")
plt.scatter(xs, ys5, c='m')
plt.plot(xFit, func(xFit, *popt5), "m", label="f = 2500Hz")
plt.scatter(xs, ys6, c='y')
plt.plot(xFit, func(xFit, *popt6), "y", label="f = 3000Hz")

plt.xlabel(r"$V_{sig}$ $(\mu V)$")
plt.ylabel(r"$V_{DC}$ (mV)")
plt.legend()
```

```python
std1, std2, std3, std4, std5, std6 = np.std(func(xs, *popt1)-ys1), np.std(func(xs, *popt2)-ys2), np.std(func(xs, *popt3)-ys3), np.std(func(xs, *popt4)-ys4), np.std(func(xs, *popt5)-ys5), np.std(func(xs, *popt6)-ys6)
std1, std2, std3, std4, std5, std6
```

```python
x = np.array([500, 800, 1200, 1600, 1800])
y = np.array([101.5, 155.6, 356.8, 453.8, 520.8])
popt, pcov = curve_fit(func, x, y)

# RESULT OUPUT
p = 3
text = "If the Equation is y = mx + c\nFor the best fit curve:\n"
text += f"    m = {round(popt[0], p)} ± {round((pcov[0][0])**0.5, p)}\n"
text += f"    c = {round(popt[1], p)} ± {round((pcov[1][1])**0.5, p)}\n"
print(text)
print("The curve looks like:")

# Just automating some stuffs....
span = max(x) - min(x)
a = 0.05 # This variable stores the amount of free space to be shown before the least point and 
# after the maximum point. Keep this value low for best results. 
factor = span*a

xFit = np.arange(min(x)-factor, max(x)+factor, 0.01) # np.arrange(start, stop, step), makes an array of numbers (range is chosen to match x data)
 
# Plotting the fout things... data points, fit, upper limit of error, lower limit of error
plt.plot(x, y, 'go') # go means green dots
 
plt.plot(xFit, func(xFit, popt[0], popt[1]), 'r')

# Comment out the next two lines if you want to hide error lines 
plt.plot(xFit, func(xFit, popt[0]+((pcov[0][0])**0.5), popt[1]+((pcov[1][1])**0.5)), "b--") # "b--" means blue dashed lines
#                                ^                            ^
plt.plot(xFit, func(xFit, popt[0]-((pcov[0][0])**0.5), popt[1]-((pcov[1][1])**0.5)), "b--")
#                                ^                            ^
 
plt.legend(["Data Points", "Least Square Fit", "Error"])
plt.xlabel(r"Frequency (f) (Hz)")
plt.ylabel(r"$slope\times10^3$ $(\alpha)$")
plt.text(0.5, 0.15, text, fontsize=12, transform=plt.gcf().transFigure)
```
